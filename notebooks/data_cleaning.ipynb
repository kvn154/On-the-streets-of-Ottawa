```python
# copying the above in a dataframe and changing last days of months by first days of next months
mw_raw_data = pd.DataFrame({
    'date': ['01-Jan-95', '01-Feb-04', '01-Feb-05', '01-Feb-06', '01-Feb-07', '01-Apr-08', '01-Apr-09', '01-Apr-10', 
             '01-Jun-14', '01-Oct-15', '01-Oct-16', '01-Oct-17', '01-Jan-18', '01-Oct-20', '01-Oct-21', '01-Jan-22', 
             '01-Oct-22', '01-Oct-23'],
    'new_wage': [6.85, 7.15, 7.45, 7.75, 8.00, 8.75, 9.50, 10.25,
                 11.00, 11.25, 11.40, 11.60, 14.00, 14.25, 14.35, 15.00,
                 15.50, 16.55]
})
mw_raw_data.head(3)
```

```python
mw_raw_data['date'] = pd.to_datetime(mw_raw_data['date'])
mw_raw_data
```

```python
mw_raw_data.iloc[17]
```

```python
mw = pd.DataFrame({
    # generating months column from Jaunuary 2001 til October 2022
    'month' : pd.date_range("2000-12-31", periods=262, freq="M"),
    'hourly_wage': [6.85 for i in range(262)]
})
mw.head(3)
```

```python
# automatic generation could only work with last days of months, so I started the month before and now add a day for each date
mw['month'] = mw['month'] + datetime.timedelta(days=1)
mw
```

```python
mw.iloc[1,0].strftime('%Y-%m')
```

```python
new_wage_idx = 1
for i in range(1,262):
    if new_wage_idx <= 17 and mw.iloc[i,0].strftime('%Y-%m') == mw_raw_data.iloc[new_wage_idx, 0].strftime('%Y-%m'):
            mw.iloc[i,1] = mw_raw_data.iloc[new_wage_idx, 1]
            new_wage_idx += 1
    else:
        mw.iloc[i,1] = mw.iloc[i-1,1]
mw
```

```python
# assuming there are 37.5 hours of work per week
mw['weekly_wage'] = mw['hourly_wage']*37.5
mw['monthly_wage'] = mw['weekly_wage']*4.33
mw
```

```python
mw.to_pickle('../clean_datasets/ontario_min_wage.pkl')
```

### Criminal offences

```python
def compute_perc(size, name, year):
    t = full_data[(full_data.Name == name) & (full_data.YEAR == year)]["size"]
    if len(t) == 0:
        return 0
    return size / t.iloc[0]  
```

```python
criminal_data_set = pd.read_csv('../datasets/Criminal_Offences_Open_Data.csv')
neighbourhood_loc = gpd.read_file("../datasets/OPS_Neighbourhoods_Open_Data.geojson")
neighbourhood_loc = neighbourhood_loc.rename(columns ={'namese2016':'Name', "namesd2016" : "NameF"})
police_centers = pd.read_csv("../datasets/Community_Police_Centres_Open_Data.csv")
neighbourhood_loc["lat"] = neighbourhood_loc.centroid.y
neighbourhood_loc["long"] = neighbourhood_loc.centroid.x
neighbourhood_dic = neighbourhood_loc.set_index("Name").geometry
full_data = criminal_data_set.groupby(['NB_NAME_EN', 'YEAR'], as_index=False).size()
full_data_by_cat = criminal_data_set.groupby(['NB_NAME_EN', 'YEAR', 'OFF_CATEG'], as_index=False).size()
full_data = full_data.merge(neighbourhood_loc, left_on = "NB_NAME_EN", right_on = "Name", how='left')

full_data_by_cat = full_data_by_cat.merge(neighbourhood_loc, left_on = "NB_NAME_EN", right_on = "Name", how='left')

full_data_by_cat["perc_of_crimes"] = full_data_by_cat.apply(lambda x: compute_perc(x["size"], x["NB_NAME_EN"], x["YEAR"]), axis = 1)
full_data_by_cat = full_data_by_cat.rename(columns ={'size':'number_of_crimes'})
full_data = full_data.rename(columns ={'size':'number_of_crimes'})

full_data.to_pickle("../clean_datasets/Criminal_Offences_clean_full.pkl")
full_data_by_cat.to_pickle('../clean_datasets/Criminal_Offences_clean_by_categ.pkl')
full_data_by_cat.to_csv('../clean_datasets/Criminal_Offences_clean_by_categ.csv')
neighbourhood_loc.to_pickle('../clean_datasets/Neighbourhood.pkl')
```

```python
full_data = pd.read_pickle("../clean_datasets/Criminal_Offences_clean_full.pkl")
traj_groups = pd.read_csv("../Trajectories/traj_nei_results.csv")
full_data_traj = full_data.merge(traj_groups, left_on = "NB_NAME_EN", right_on = "NB_NAME_EN", how='left')
full_data_traj.to_pickle("../clean_datasets/Criminal_Offences_clean_full_traj.pkl")
```

### Confirmed Opioid related Deaths of Ottawa Residents

Opioid-related deaths are **defined as an acute intoxication/toxicity death resulting from the direct effects of the administration of exogenous substance(s) where one or more of the substances is an opioid**, regardless of how the opioid was obtained. This excludes deaths due to chronic substance use, medical assistance in dying, trauma where an intoxicant contributed to the circumstances of the injury and deaths classified as homicide.

According to this study https://www.canada.ca/en/health-canada/services/opioids/data-surveillance-research/homelessness-substance-related-acute-toxicity-deaths.html opioid related deaths are commonly refered to as overdoses

```python
overdose_related_deaths=pd.read_csv('../datasets/Confirmed_Opioid_related_Deaths_of_Ottawa_Residents_.csv')
overdose_related_deaths= overdose_related_deaths.drop('ObjectId', axis=1)
overdose_related_deaths['Year'] = overdose_related_deaths['Quarter'].str.split(' - ', expand=True)[0]
overdose_related_deaths['Quarter'] = overdose_related_deaths['Quarter'].str.split(' - ', expand=True)[1]
overdose_related_deaths
```

```python
overdose_related_deaths.to_csv('../clean_datasets/overdose_related_deaths.csv', index=False)
```
